## NBA Analysis

```python
import pandas as pd
```

```python
nba = pd.read_csv("./pandas/nba.csv")
nba
```

```python
nba.head(10)
nba["Name"].value_counts()
len(nba.index)
```

```python
avg_salary = nba["Salary"].mean()
round(avg_salary, 2)
highest = nba["Salary"].max()
```

```python
nba[nba["Salary"] == highest][["Name","Team"]]
```

```python
age_filter = (20 <= nba["Age"]) & (nba["Age"] <=25)
nba[age_filter][["Name","Team","Age"]].sort_values(["Age", "Name"],ascending=[False,True])
```

```python
name_filter = nba["Name"] == "John Holland"
nba[["Name","Team"]][name_filter]
nba["Team"][name_filter].iloc[0]
```

```python
group_by_team = nba.groupby("Team")["Salary"].mean()
group_by_team.sort_values(inplace=True, ascending=False)
round(group_by_team)
```

```python
# team count
nba["Team"].nunique()
# players by team
nba["Team"].value_counts()
```

```python
nba.dropna(how="any",inplace=True)
and_filter = nba["Name"].str.lower().str.contains("and")
nba[and_filter]
```

```python
youtube = pd.read_csv("./datasets/youtube-ing.csv")
youtube.head()
youtube.info()
```

```python
youtube.head(10)
youtube.iloc[10:15]
```

```python
youtube.describe()
youtube.columns
len(youtube.columns)
# youtube.value_counts
```

```python
# drop some columns
youtube.drop(labels=["thumbnail_link","comments_disabled","ratings_disabled","video_error_or_removed","description"], axis=1, inplace=True)
youtube
```

```python
youtube[['likes', 'dislikes']].mean()
youtube.columns
```

```python
youtube[:50][['likes', 'dislikes']]
youtube.head(50)[['likes', 'dislikes']]
```

```python
max_views = youtube['views'] == youtube['views'].max()
min_views = youtube['views'] == youtube['views'].min()
youtube[max_views]
youtube[min_views]
```

```python
youtube.sort_values("views", ascending=False).head(10)[["title","views"]]
```

```python
like_by_category = youtube.groupby('category_id')['likes'].mean()
like_by_category.sort_values(ascending=False)
```

```python
comments_by_category = youtube.groupby('category_id')['comment_count'].sum()
# comments_by_category
comments_by_category.sort_values(ascending=False)
```

```python
youtube["category_id"].value_counts()
```

```python
youtube["Title Length"] = youtube["title"].apply(len)
youtube
```

```python
youtube["Tag Count"] = youtube["tags"].apply(lambda num:len(num.split("|")))
youtube
```

```python
# do not include zor like or dislikes!
youtube["popularity"] = youtube["likes"] / youtube["dislikes"]
youtube.sort_values("popularity",ascending=False)[youtube["dislikes"]>0].head(10)
```
