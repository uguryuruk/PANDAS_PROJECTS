```python
country=pd.read_csv("./pandas/dunya.csv")
country
country.describe()
country.info()
```

```python
country["Ülke"]
country[["Ülke","Gelir"]]
country["Gelir"]
country["KITA"]
country[0:5]["KITA"]
```

```python
million_filter = country["Gelir"] < 1000000
income_filter = country["Kişi Başına"] < 10000
continent_filter = country["KITA"].isin(["Avrupa","Asya"])
country[million_filter & income_filter]
country[continent_filter]
country[million_filter]
country[income_filter]
country[million_filter | income_filter]
```

```python
country["Gelir TL (bin)"] = country["Gelir"] * 28 /1000
country["Uzun Kıta"] = country["KITA"] + " Kıtası"
country
country["KITA"].value_counts()
```

```python
country["Nüfus"]= country["Nüfus"].str.replace(",","")
country["Nüfus"] = country["Nüfus"].astype(int)
```

```python

country.sort_values(["KITA", "Nüfus"],ascending=False)
```

```python
continent = country.groupby("KITA")[["Gelir","Kişi Başına"]].mean()
continent
```
