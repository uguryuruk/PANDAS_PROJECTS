SAMPLE, NSMALLEST, QUERY, APPLY

```python
takim_milli=pd.read_csv("./datasets/millitakim.csv",index_col="İsim")
takim_milli
```

```python
# rastgele örnek için sample
takim_milli.sample(5,axis=0)
```

```python
# en küçük n değer

takim_milli.nsmallest(3,columns="Yaş")

# en büyük n değer

takim_milli.nlargest(3,columns="Yaş")
```

```python
fener=takim_milli["Takım"]=="Fenerbahçe"
# diğerlerini NaN yaparak verir.
takim_milli.where(fener)
```

```python
yas=takim_milli["Yaş"]<30
# hem fenerli hem yaşı 30 dan düşükler:
takim_milli.where(fener & yas)
```

```python
takim_milli.query('Takım=="Fenerbahçe"')
takim_milli.query('Görev == "sol bek"')
takim_milli.query('Yaş < 25 or Görev == "sol bek"')
```

```python
# apply için fonksiyon yazmalıyız, onu kullanmamızı sağlıyor, her değeri işlemeyi sağlar map gibi.

def milyon(number):
    return str(number) + " milyon euro"

sutun= ["Milyon Euro"]
for sut in sutun:
    takim_milli[sut]= takim_milli[sut].apply(milyon)

takim_milli
```

```python
takimlar=takim_milli["Takım"].copy()
takimlar["Harun Tekin"]="Türkiye"
takimlar
```

```python
x.loc["2014",2]
```

```python
# depracated olabilir, kontrol et.
on.append(alti,ignore_index=True)
```
