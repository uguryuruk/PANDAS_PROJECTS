add age column calculated by this year
(HINT: ask chatgpt; Hello, I am using python pandas framework and a column is birthday format, which is in string format like "03.04.1990". I want to calculate the ages in a new column. How can I do that?)

```python
# TODO: search for changing string to date (maybe year)
from datetime import datetime
year = datetime.now().year
this_date = datetime.now()
world_cup["Age"] = pd.to_datetime(world_cup["Birth Date"])
# world_cup["Age"] = world_cup["Age"].apply(lambda x: year - x.year)
world_cup["Age"] = this_date - world_cup["Age"]
world_cup["Age"] = world_cup["Age"].apply(lambda x: x.days // 365)
```
